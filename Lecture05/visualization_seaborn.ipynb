```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset('tips')
tips.head()

### bar plot
# Bar 플롯
plt.figure(figsize=(5,3))
sns.barplot(x='day', y='tip',data=tips)
```

```python
### scatter plot ()
# 흩어지는 플롯
plt.figure(figsize=(3,3))
sns.jointplot(x='total_bill', y='tip', data=tips)

plt.figure(figsize=(3,3))
sns.jointplot(x='size', y='tip', data=tips)
```

```python
### histogram 표 
plt.figure(figsize=(5,5))
sns.distplot(tips['tip'], bins=30)
plt.show()
```

```python
### box plot
sns.boxplot(x='day', y='tip',data=tips)
plt.show()

sns.boxplot(data=tips, x='day', y='tip', hue='smoker')
plt.show()

sns.boxplot(data=tips, x='day', y='tip', hue='sex')
plt.show()
```

```python
### line plot
sns.lineplot(x=tips['total_bill'], y=tips['tip'])
```

```python
### regression plot 회귀직선
sns.regplot(x='total_bill', y='tip', data=tips)
```

```python
### countplot
sns.countplot(x=tips.sex, data=tips)
plt.show()
```

```python
### titanic
titanic = sns.load_dataset('titanic')
titanic.head()
# 남성과 여성의 수를 bar chart로 표현하시오..
sns.countplot(x=titanic.sex, data=titanic)
sns.countplot(x=titanic.who, data=titanic)
sns.countplot(x=titanic.who, data=titanic, hue='pclass')
```

```python
# 승객의 연령분포
sns.distplot(titanic['age'], bins=50)
```
