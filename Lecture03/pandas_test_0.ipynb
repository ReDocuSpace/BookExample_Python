```python
raw_data= {'first_name':['jason',np.nan, 'tina','jake','amy'],
          'last_name':['miller', np.nan, 'all','miler','coozer'],
          'age':[42, np.nan, 36, 24, 73],
          'sex':['m', np.nan, 'f','m','f'],
          'pretestscore':[4, np.nan, np.nan, 2, 3],
          'posttestscore':[25, np.nan, np.nan, 62, 70]}
```

```python
df = pd.DataFrame(raw_data)
df
```

```python
#36 row별 nan값의 개수를 출력하시오.
df.isnull().sum(axis=1)
```

```python
#37 column별 nan값의 개수를 출력하시오..
df.isnull().sum()
```

```python
#38 nan값이 포함된 row를 삭제하시오.
df.drop(df[df.isnull().sum(axis=1) > 0].index)
df
```

```python
#40 데이터가 2개 이하인 row를 삭제하시오
#41 데이터가 3개 이상이면 삭제하지 마시오

df.drop(df[df.isnull().sum(axis=1) > 2].index,inplace=True)
df


```

```python
#42 nan값이 포함된 column을 삭제하시오.
df.drop(df.loc[:, df.isnull().sum() != 0].columns,axis=1)
```

```python
#43 nan값을 ‘0’으로 채우시오.
df.fillna(0)
```

```python
#44 ‘age’를 기준으로 내림차순으로 정렬하시오.
df.sort_values('age',ascending=False)
```

```python
#45 ‘sex’ 칼럼에서 남성이면 0, 여성이면 1의 값으로 수정하시오..

df['sex'].replace({'f' : 1, 'm' : 0})
```
