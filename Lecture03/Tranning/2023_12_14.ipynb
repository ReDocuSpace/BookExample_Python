```python
### pandas Visualization
url = r'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv'
drinks = pd.read_csv(url)
drinks.head()

drinks.info()

drinks.columns = ['country', 'beer', 'spirit', 'wine', 'litres', 'continent']
drinks.head(3)

drinks.set_index('country', inplace=True)
drinks.head()

plt.figure(figsize=(10,3))
drinks.beer.sort_values().plot()

plt.figure(figsize=(10,3))
drinks.beer.sort_values()[-10:].plot()

plt.figure(figsize=(10,3))
drinks.plot()
```

```python
# Bar Chart
drinks['beer'].sort_values(ascending=False).iloc[:20].plot(kind='bar', figsize=(10,3))
```

```python
# 대륙별 맥주 소비량을 bar chart로 나타내시오
drinks.groupby('continent')['beer'].sum().plot(kind='bar', figsize=(10,3))

```
