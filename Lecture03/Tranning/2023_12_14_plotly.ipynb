```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import plotly.express as px

df = px.data.stocks()
df.head(5)
```

```python
#px.line
px.line(df, x='date', y='GOOG')
```

```python
df1 = df.set_index('date')
df1.head(3)
```

```python

px.line(df, x='date', y=['GOOG', 'AAPL', 'FB'])
px.bar(df1, x=df1.index, y='GOOG')

# facet_col

df1.columns.name = 'company'
px.area(df1, facet_col='company', facet_col_wrap=2)
df1.columns.name = 'company'
px.area(df1, facet_col='company', facet_col_wrap=3)

fig = px.line(df, x='date', y=df.columns[:3], hover_data={'date':"|%Y-%m_d"})
fig.update_xaxes(dtick='D5')    # x  - 5

fig = px.line(df, x='date', y=df.columns[:3], hover_data={'date':"|%Y-%m_d"})
fig.update_xaxes(dtick='D5')    # x  -  
```

```python
import plotly.graph_objects as go

df = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/finance-charts-apple.csv')
df.head()

fig = go.Figure(data=[go.Candlestick(x=df['Date'],
                                     open=df['AAPL.Open'],
                                     high=df['AAPL.High'],
                                     low=df['AAPL.Low'],
                                     close=df['AAPL.Close'])])
fig.show()
```

```python
import FinanceDataReader as fdr

amd = fdr.DataReader('AMD', '2015', '2021')
amd.head()
```

```python
amd['Change'] = amd['Close'].pct_change()

px.histogram(amd, x='Change', marginal='violin', title='')

FAANG = ['META', 'AMZN', 'AAPL', 'NFLX', 'GOOGL', 'MSFT']

faang_list = [fdr.DataReader(code, '2019', '2024')['Close'] for code in FAANG]

fa_df = pd.concat(faang_list, axis=1)
fa_df.columns = FAANG
fa_df.head()
df_ratio = fa_df/fa_df.iloc[0]
```

```python
df_ratio.columns.name = 'company'
px.line(df_ratio, title='FAANG  ')

px.area(df_ratio, facet_col='company', facet_col_wrap=3)
```

```python

px.scatter(df_ratio, x='GOOGL', y='MSFT')
px.box(df_ratio)
```

```python
px.histogram(df_ratio, facet_col='company', facet_col_wrap=2)
```
