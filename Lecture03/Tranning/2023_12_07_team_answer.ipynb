```python
import numpy as np
import pandas as pd

age_pd = pd.read_csv('eda_data_files\\age_pd.csv',encoding='euckr')

age_pd.head()
age_pd.columns

age_pd.drop(['2019년02월_계_총인구수', '2019년02월_계_연령구간인구수'],axis = 1,inplace=True)
age_pd.drop(0, inplace=True)
age_pd.head()

age_pd.columns = pd.Series(age_pd.columns).apply(lambda x:x.split('_')[-1])
#age_pd.iloc[0]

age_pd.index = age_pd['행정구역'].apply(lambda x : x.split()[1])
age_pd.drop(['행정구역'],axis=1,inplace=True)
age_pd.head()

for col in age_pd.columns:
    try:
        age_pd[col] = age_pd[col].apply(lambda x : x.replace(',','')).astype('int')
    except:
        pass
    
age_pd
    
```

```python
#5

def cal_g_pop(g):

  gen = str(g)[0]

  if g == 0:
    s_g = age_pd.loc[:, '0세':'9세'].sum(axis=1)
  elif g==100:
    s_g = age_pd.loc[:, '100세 이상']
  else:
    s_g = age_pd.loc[:, '{}0세'.format(gen):'{}9세'.format(gen)].sum(axis=1)

  return s_g

result = cal_g_pop(0)
result.sort_values().index[-1]
```

```python
#6

for g in range(0,100,10):
    age_pd['{}대'.format(g)] = cal_g_pop(g)
    
df_gen = age_pd.loc[:, '0대':]
df_gen_cor = df_gen.T.corr()
df_gen_cor['종로구'].sort_values(ascending=False)[:3]
```
