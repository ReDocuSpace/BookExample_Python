```python
# 엑셀파일 읽기
##!pip install xlrd

import numpy as np
import pandas as pd

cctv = pd.read_csv('D:\BookExample_Python\Lecture03\Tranning\Backup\eda_data_files\CCTV_in_Seoul.csv')

cctv.rename(columns={'기관명':'구별'}, inplace=True)
cctv.rename(columns={'소계' : '총CCTV수'}, inplace=True)

cctv.info()

people = pd.read_excel('D:\BookExample_Python\Lecture03\Tranning\Backup\eda_data_files\population_in_Seoul.xls')

people.drop(['기간','세대','세대당인구'], axis= 1, inplace= True)
people.columns = ['구별','총인구','총남인구','총여인구','한계','한국남성','한국여성','외국인','남외국인','여외국인','고령자']
people.drop([0,1,2], axis=0, inplace= True)
people

for col in people.columns[1:]:
    people[col] = pd.to_numeric(people[col])

people.info()
```

```python
# CCTV가 가장 많은 구는..?
cctv.sort_values('총CCTV수', ascending=False).iloc[0]['구별']
```

```python
# 2016년에 cctv증가율이 가장 높은 구는?
cctv['2016년증가율'] = (cctv['2016년'] - cctv['2015년']) / cctv['2015년']
cctv[cctv['2016년증가율'].max() == cctv['2016년증가율']]

max_idx = cctv['2016년증가율'].argmax()
cctv.iloc[max_idx]
```

```python
# 외국인 비율이 가장 높은 구는?
people['외국인비율'] = people['외국인'] / people['총인구']
people.sort_values('외국인비율', ascending=False).iloc[0]
```

```python
# 고령자 비율이 가장 높은 구는?
people['고령자비율'] = people['고령자'] / people['총인구']
people.sort_values('고령자비율', ascending=False).iloc[0]
```

```python
# 전체 인구수가 가장 많은 구는?
people.sort_values('총인구', ascending=False).iloc[0]
```

```python
# 서울시 인구당 cctv수가 많은 구...(인구대비 cctv비율)

people.dropna(inplace=True)

temp = cctv['총CCTV수'].reset_index(drop=True) / people['총인구'].reset_index(drop=True) 
temp.index = people['구별']
```

```python
merged_df = pd.merge(people,cctv,on="구별")

merged_df['인구당cctv수'] = merged_df['총CCTV수'] / merged_df['총인구']
merged_df.sort_values('인구당cctv수', ascending=False).iloc[:5]
```
