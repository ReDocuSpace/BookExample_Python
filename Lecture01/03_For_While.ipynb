```python
L1 = ['A', 'B', 'C']
L2 = [1, 2, 3]

# L1과 L2의 원소 간의 모든 조합을 만들어서 리스트에 추가하시오 (2중 for문)
# A1, A2, A3, ... C2, C3

L3 = []

for i in L1:
  for j in L2:
    L3.append(i+str(j))

L3
```

```python
words = ['mountains', 'sea', 'sky', 'river', 'cloud', 'sea', 'sky', 'sea','mountains']
# 위의 단어 중에서 unique한 단어로만 리스트를 구성하시오.(중복단어 제거) 단, set()은 사용하지 마시오...

L = []

for w in words:
  if w not in L:
    L.append(w)
L
```

```python
# 1~50까지 정수에서 홀수의 합과 짝수의 합을 각각 구하시오..

even_sum = 0
odd_sum = 0

for i in range(51):
  if i%2 == 0:
    even_sum += i
  else:
    odd_sum += i

print(even_sum, odd_sum)
```

```python
words = ['mountains', 'sea', 'sky', 'river', 'cloud', 'sea', 'sky', 'sea','mountains']

# 위의 단어에서 단어의 길이가 4 이상인 단어만 중복없이 저장(출력)하시오..

s = set()

for w in words:
  if len(w) >= 4:
    s.add(w)

s
```

```python
### while 조건식:
#      실행코드
#      실행코드
# ...

# 조건식이 True이면 실행코드를 반복을 지속,
# 조건식이 False가 되면, 반복을 멈춤...

# 조건식이 계속 True이면 실행코드가 무한반복 -  무한루프
# while문을 사용 시 조건식이 False가 되어서 반복을 머추는 장치를 만들어야 한다..
```

```python
import time

while True:
  print('python', end=' ')
  time.sleep(1)
```

```python

a = 1
suma = 0

while a <= 10:
  suma += a
  a += 1  # 무한반복을 중단하는 장치

suma
```

```python
a = 1
suma = 0

while a <= 10:
  if a%2 == 0:
    suma += a
  a += 1  # 무한반복을 중단하는 장치

suma
```

```python
# 반복문에서 break를 만나면 반복문을 강제적으로 빠져나간다
```

```python
a = 1
suma = 0

while a <= 10:
  print(a, end=' ')
  a += 1  # 무한반복을 중단하는 장치
```

```python
a = 1
suma = 0

while a <= 10:
  if a >= 5:
    break   # 반복문에서 break가 실행되면 반복문을 빠져나간다.
  print(a, end=' ')
  a += 1  # 무한반복을 중단하는 장치
```

```python
text = "python is fun!!"

# text에서 문자을 하나씩 출력하다가 space를 만나면 반복문을 빠져나가도록 하시오

i = 0
while True:
  print(text[i], end='')
  i += 1
  if text[i] == ' ':
    break

```

```python
# continue : 반복문에서 continue가 실행되면, 반복처음으로 돌아간다..
#            continue 아래 코드를 실행하지 않는다.
```

```python
a = 0

while a <= 10:
  a += 1
  if a%3 == 0:
    continue
  print(a)
```

```python
# 0~10까지 정수를 반복하는 반복문에서 0~5까지만 출력되도록 하시오, while문과 break를 사용하시오..
```

```python
count = 0
while count <= 10:
  if count > 5:
    break
  print(count)
  count += 1
```

```python
# 0~10까지 정수를 반복하는 반복문에서 5이상만 출력되도록 하시오, while문과 continue를 사용하시오..

count = 0
while count < 10:
  count += 1
  if count < 5:
    continue
  print(count)
```

```python
# 0~10까지 정수를 반복하는 반복문에서 3,4,5,6만 출력되도록 하시오, while문과 break, continue를 사용하시오..
```

```python
count = 0

while count <= 10:
  count += 1

  if count < 3:
    continue
  if count > 6:
    break

  print(count)
```

```python
# 점수를 5번 입력을 받은 후, 점수의 평균을 출력하시오.
# 단, while문을 사용하시오...

suma = 0
count = 0

while count < 5:
  score = int(input('점수를 입력하세요..'))
  suma += score
  count += 1

total_sum = suma/count
print(total_sum)
```
