```python
# 체중과 키를 입력받은 후, bmi를 계산해서 30이상이면 "비만입니다"
# 30 ~ 25 사이이면 "과체중입니다"
# 25 ~ 18 사이이면 "정상입니다"
# 18 미만이면 "저체중입니다"를 출력하시오
# bmi = weight/height**2

weight = int(input("체중을 입력해주세요 : "))
height = int(input("키를 입력해주세요 : "))

bmi = (weight/((height*0.01) ** 2))

print(bmi)

if bmi >= 30:
    print("비만입니다")
elif bmi >= 25 and bmi < 30:
    print("과체중입니다")
elif bmi >= 18 and bmi < 25:
    print("정상입니다")
elif bmi < 18:
    print("저체중입니다.")
else :
    print("정상적이지 않은 값입니다")
```

```python
# 주민번호를 앞 7 자리를 입력받은 후
# 40세 이상이면 중년
# 20세 이상이면 청년
# 20세 미만이면 청소년

#"당신은 < 남성/여성> 이고, <중년/청년/청소년>입니다"

person_id = input()

year_check = int(person_id[:2])
gender_check = int(person_id[-1:])

year = ""
gender = ""
current_year = 2023

age = current_year - (year_check + 1900)

if age > 40:
    year = "중년"
elif age > 20:
    year = "청년"
else:
    year = "청소년"

if gender_check % 2 == 0:
    gender = "여성"
else:
    gender = "남성"

print(f"당신은 {gender} 이고, {year}입니다")
```

```python
# 이메일을 입력받은 후 "admin@gmail.com"과 비교해서
# 아이디만 틀리면 "아이디가 틀립니다"
# 도메인만 틀리면 "도메인이 틀립니다"
# 아이디와 도메인 모두 틀리면 "아이디와 도메인이 모두 틀립니다"
# 아이디와 도메인이 일치하면 "이메일이 일치합니다.환영합니다"를 출력하시오

email = input()

email_id = email.split("@")[0]
email_domain = email.split("@")[1]

root_id = "admin"
root_domain = "gmail.com"

# 다중 조건
if email_id != root_id and email_domain == root_domain:
    print("아이디가 틀립니다")
elif email_id == root_id and email_domain != root_domain:
    print("도메인이 틀립니다")
elif email_id != root_id and email_domain != root_domain:
    print("아이디와 도메인이 모두 틀립니다")
elif email_id == root_id and email_domain == root_domain:
    print("이메일이 일치합니다 환영합니다.")

# 중첩 조건문
if email_id == 'admin':
    if email_domain == 'gmail.com':
        print("이메일이 일치합니다 환영합니다.")
    else:
        print("도메인이 틀립니다")
else:
    if email_domain == 'gmail.com': 
        print("아이디가 틀립니다")
    else:
        print("아이디와 도메인이 모두 틀립니다")
```

```python
# 임의의 정수 a와 b를 입력받은 후,
# a가 b보다 크면 'a가 b보다 큽니다'
# a가 b보다 작으면 'a가 b보다 작습니다'
# a와 b가 같으면 'a 와 b가 같습니다'를 출력하시오

a = int(input())
b = int(input())

if a > b:
    print(f"{a}가 {b}보다 큽니다")
elif a < b:
    print(f"{a}가 {b}보다 작습니다")
else:
    print(f"{a}와 {b}가 같습니다")
```
