```python
solarsys = ['태양', '수성', '금성', '지구', '화성', '목성', '지구']
solarsys1 = ['토성', '천왕성', '해왕성']
planet = '지구'
L1 = [2,7,4,9,5,3]
```

```python
# solarsys 에서 #   ['수성', '금성', '지구']를 출력하시오
solarsys[1:4]
# '지구'의 인덱스값 출력
solarsys.index('지구')
# '화성'을 '해왕성'으로 수정하시오
solarsys[solarsys.index('화성')] = '행왕성'
# 맨 뒤에 '토성'을 추가하시오
solarsys.append('토성')
# '수성'자리에 '명왕성'을 삽입
solarsys.insert(1, '명왕성')
solarsys.insert(solarsys.index('수성'), '명왕성')
# 맨 끝 요소를 삭제하고 삭제한 값을 리턴하시오
solarsys.pop()
# 1 ~ 3 번 인덱스까지 삭제하시오
del solarsys[1:4]
# solaysys1에서 천왕성을 삭제하시
solarsys1.remove('천왕성')
#  인덱스 2번 요소를 삭제하시오
solarsys.pop(2)

# solarsys와 solarsys1을 합하시오
solarsys + solarsys1
solarsys.extend(solarsys1)
# solarsys 의 원소의 순서를 역순으로 하시오
solarsys.reverse()
# solarsys에 '만리장성'이 있는지 확인하시오
'만리장성' in solarsys
# solarsys1을 세 번 반복하시오
solarsys * 3
# solaysys의 길이를 출력하시오
len(solarsys)
# solarsys의 모든 원소를 지우시오 - 빈리스트
solarsys.clear()
# solarsys 리스트를 제거하시오
del solarsys
# L1의 합, 최소값, 최대값, 평균값을 출력하시오..
sum(L1); min(L1), max(L1), sum(L1)/len(L1)

import numpy as np
np.mean(L1)
```

```python
# set quiz

# 원소가 없는 set인 s를 만드세요
# s = {} - dict
s = set()
# s에 원소 'sky'와 'earth'를 추가하세요
s.add('sky')
s.add('earth')
# s에 1,2 두 원소를 추가하세요
s.add(1)
s.add(2)
# s에서 'sky'를 삭제하세요
s.remove('sky')
# s.discard('sky')

A = {1,2,3,4,5,6}
B = {1,2,3,7,8,9}
C = {1,2,3}
# 두 집합의 합집합을 출력
A.union(B)
# 두 집합의 교집합을 출력
A.intersection(B)
# A의 B에 대한 차집합
A.difference(B)
# 대칭차집합
A.symmetric_difference(B)
# C가 A의 부분집합인지 확인
C.issubset(A)
# A가 C를 포함하는지 확인
A.issuperset(C)

# 집합 B의 길이를 출력
len(B)
# 집합 B의 최대값 출력
max(B)
# 집합 B의 최소값 출력
min(B)
# 집합 B의 합을 출력
sum(B)
# 집합 B의 평균출력
import numpy as np
np.mean(B)

sum(B)/len(B)

# 집합 B의 내림차순 출력
sorted(B, reverse=True)
# 집합 B에 5가 포함되어 있는지 확인
5 in B
```
