```python
# set_info()를 실행하지 않고 businesscard()을 실행하면 인스턴스 값이 할당되지 않아서 error
# => 인스턴스 생성 시 필요한 변수에 초기값을 할당하자
# => 생성자함수() : 생성자 함수는 인스턴스생성 시 자동 실행 => 강제로 변수에 초기값을 설정하도록 할 수 있음.
```

```python
class BusinessCard2:

  Name = 'Kim'    # 클래스 변수
  Email = 'kim@hanmail.net'
  Address = 'Seoul'

  # 생성자 함수 - 인스턴스 생성 시 자동 실행..
  def __init__(self, name, email, address):  # 생성자 함수..
    self.name = name  # self.name : 인스턴스 변수
    self.email = email
    self.address = address

  def businesscard(self):  # 파이썬에서 메소드에는 self매개변수를 하나 가진다..
    print("="*20)
    print("Name : {}".format(self.name))
    print("Email : {}".format(self.email))
    print("Address : {}".format(self.address))
    print("="*20)
```

```python
class Circle:

  pi = 3.14  # 클래스 변수 - 모든 인스턴스에서 동일한 값을 공유, 변하지 않는 값

  def __init__(self, r):  # 생성자함수 - 인턴스 생성 시 자동 실행...
    self.radius = r  # 인스턴스 변수 - 인스턴스마다 다른 값을 가질 수 있다..

  def area(self):
    circle_area = Circle.pi * self.radius**2
    return circle_area

  def perimeter(self):
    circle_perimeter = Circle.pi * self.radius*2
    return circle_perimeter
```

```python
# 클래스 : FourCal
# 인스턴스 변수 a, b - 정수타입
# 사칙연산 메소드 - 더하기메소드, 빼기 메소드, 곱셈 메소드, 나눗셈 메소드

# 인스턴스를 생성후 각각의 메소드를 호출해서 실행..
```

```python
class FourCal:
  def __init__(self, a, b):
    self.a = a
    self.b = b
  def addf(self):
    result = self.a + self.b
    return result
  def minf(self):
    result = self.a - self.b
    return result
  def mult(self):
    result = self.a*self.b
    return result
  def divf(self):
    result = self.a/self.b
    return result
```

```python
fc = FourCal(2,4)
```

```python
fc.addf()
```

```python
fc.divf()
```

```python
class FourCal2(FourCal):

  def divf(self):  # 부모 클래스 메소드 수정 => override
    if self.b == 0:
      return 0
    else:
      return self.a/self.b

  def powf(self):  # 추가 메소드
    result = self.a**self.b
    return result
```

```python
students = [
    {'name':'윤길수', 'korean':87, 'english':95, 'science':87},
    {'name':'임철수', 'korean':87, 'english':95, 'science':87},
    {'name':'박민호', 'korean':87, 'english':95, 'science':87},
    {'name':'이수창', 'korean':87, 'english':95, 'science':87},
    {'name':'강철수', 'korean':87, 'english':95, 'science':87},
]
```

```python
print("Name", "Suma", "Avg")

for s in students:
  name = s['name']
  kor = s['korean']
  eng = s['english']
  sci = s['science']

  suma = kor + eng + sci
  avg = suma / 3
  print(name, suma, round(avg))
```

```python
def suma(students):
  for s in students:
    kor = s['korean']
    eng = s['english']
    sci = s['science']

  suma = kor + eng + sci
  return suma
```

```python
suma(students)
```

```python
def avgf(students):
  for s in students:
    kor = s['korean']
    eng = s['english']
    sci = s['science']

  res = (kor + eng + sci)/3
  return res
```

```python
avgf(students)
```

```python
class Student:
  def __init__(self, name, kor, math, eng, sci):
    self.name = name
    self.kor = kor
    self.math = math
    self.eng = eng
    self.sci = sci

  def get_sum(self):
    return self.kor + self.math + self.eng + self.sci

  def get_avg(self):
    return self.get_sum()/4

  def to_string(self):
    return "{}\t{}\t{}".format(self.name, self.get_sum(), self.get_avg())
```

```python
ins = [
    Student("윤인성", 89, 78,76, 98),
    Student("박인성", 89, 78,76, 98),
    Student("강인성", 89, 78,76, 98),
    Student("임인성", 89, 78,76, 98),
    Student("이인성", 89, 78,76, 98)]

print("{}\t{}\t{}".format("Name", "Sum", "Avg"))

for s in ins:
  print(s.to_string())
```

```python
# 데이터 은닉화

import math

class Circle:
  def __init__(self, r):
    self.__radius = r  # class외부에서 접근 차단
  def get_area(self):
    return math.pi * self.__radius**2
  def get_perimeter(self):
    return math.pi * self.__radius*2

  # getter, setter
  # getter : 은닉된 변수의 값을 리턴하는 메소드 @property
  # setter : 은닉된 변수의 값을 수정하는 메소드 @get_변수명.setter

  @property
  def get_radius(self):
    return self.__radius

  @get_radius.setter
  def set_radius(self,value):
    self.__radius = value
```

```python
cir = Circle(2)
```

```python
cir.get_area()
```

```python
cir.get_perimeter()
```

```python
cir.__radius
```

```python
cir.get_radius  # getter
```

```python
cir.set_radius = 10  # setter
```

```python
cir.get_radius
```
