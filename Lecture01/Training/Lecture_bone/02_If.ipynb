```python
# 조건문 : If문
# 다양한 조건에 따라서 다르게 실행해야 하는 경우 조건문을 사용..
```

```python
# 1 case

# if 조건식:
#   실행 코드
#   실행 코드

# 조건식이 True이면 실행코드를 실행하고 False이면 아무것도 실행하지 않는다
```

```python
# 2 case

# if 조건식:
#   실행코드1
#   실행코드2
# else:
#   실행코드3
#   실행코드4

# 조건식이 True이면 실행코드1, 2를 실행하고, False이면 실행코드3, 4를 실행한다
```

```python
# 3 case

# if 조건식1:
#   실행코드1
# elif 조건식2:
#   실행코드2
# elif 조건식3:
#   실행코드3
# elif 조건식4:
#   실행코드4
# (else:
#   실행코드5)

# 조건식1이 True이면 실행코드1을 실행하고
# 조건식2가 True이면 실행코드2를 실행하고
# 조건식3이 True이면 실행코드3을 실행하고
# 조건식4가 True이면 실행코드4를 실행하고
# (조건식 1,2,3,4 모두 False이면 실행코드5를 실행한다)
```

```python
# case 1
# 성적을 입력받은 후, 성적이 80점 이상이면 "합격입니다"를 출력하시오..

score = int(input('점수를 입력하세요'))

if score >= 80:
  print("합격입니다..")
```

```python
# case 2
# 성적을 입력받은 후, 성적이 80점 이상이면 "합격입니다",
# 80점 미만이면 "불합격입니다"를 출력하시오..

score = int(input('점수를 입력하세요'))

if score >= 80:
  print("합격입니다..")
else:
  print('불합격입니다..')
```

```python
# case 3
# 성적을 입력받은 후, 성적이 80점 이상이면 "합격입니다",
# 80점 미만이면 "수고하셨습니다"를 출력하고
# 70점 미만이면 "분발하세요"를 출력하고,
# 60점 미만이면 "다음 기회에 도전하세요"를 출력하시오

score = int(input('점수를 입력하세요'))

if score >= 80:
  print("합격입니다..")
elif score >= 70:
  print("수고하셨습니다")
elif score >= 60:
  print('분발하세요')
else:
  print('다음 기회에 도전하세요..')
```

```python
# 임의의 정수를 입력받은 후 짝수이면 "짝수입니다", 홀수이면 "홀수입니다"를 출력하시오..

num = int(input('임의의 정수를 입력하세요..'))

if num%2 == 0:
  print('짝수입니다')
else:
  print("홀수입니다..")
```

```python
# 체중과 키를 입력받은 후, bmi를 계산해서 30 이상이면 "비만입니다",
# 30 ~ 25 사이이면, "과체중입니다",
# 24 ~ 18 사이이면, "정상입니다"
# 18 미만이면 "저체중입니다"를 출력하시오

# bmi = weight/((height*0.01)**2)
```

```python
weight = int(input('체중(kg)을 입력하세요'))
height = int(input('키(cm)를 입력하세요'))

bmi = weight/((height*0.01)**2)
print(bmi)

if bmi >= 25:
  print('과체중입니다..')
elif bmi >= 18:
  print('정상입니다')
else:
  print('저체중입니다..')
```

```python
# 주민번호를 앞 7자리를 입력받은 후,
#  (주민번호 앞 6자리가 생년월일, 7자리가 1이면 남성, 2면 여성)
#  (age = 2023 - 출생년도)
#  40세 이상이면 중년, 20세 이상이면 청년. 20세 미만이면 청소년으로 간주하고
# "당신은 <남성/여성>이고, <중년/청년/청소년>입니다"를 출력하시오

res_num = input('주민번호 7자리를 입력하세요..')

if res_num[-1] in ['1', '2']:
  birth_year = res_num[:2]
  age = 2023 - int('19'+birth_year)

else:
  birth_year = res_num[:2]
  age = 2023 - int('20'+birth_year)

if res_num[-1] in ['1', '3']:
  gender = '남성'
else:
  gender='여성'

if age >= 40:
  gen = '중년'
elif age >= 20:
  gen = '청년'
else:
  gen = '청소년'

text = "당신은 {}이고, {}입니다".format(gender, gen)
print(text)
```

```python
# 이메일을 입력받은 후 "admin@gmail.com"과 비교해서
# 아이디만 틀리면 "아이디가 틀립니다"
# 도메일남 틀리면 "도메인이 틀립니다"
# 아이디와 도메인 모두 틀리면 "아이디와 도메인이 모두 틀립니다",
# 아이디와 도메인이 일치하면 "이메일이 일치합니다. 환영합니다"를 출력하시오.

email = input("이메일을 입력하세요")

email_id = email.split('@')[0]
email_domain = email.split('@')[1]

if email_id == 'admin':
  if email_domain == 'gmail.com':
    print('이메일이 일치합니다. 환영합니다')
  else:
    print('도메인이 틀립니다')
else:
  if email_domain == 'gmail.com':
    print('아이디가 틀립니다')
  else:
    print('아이디와 도메인이 모두 틀립니다..')
```

```python
# names = {'park':1500, 'kim':5478, 'minho':8754, 'lee':4589}
# key를 입력받은 후, key가 names에 있으면 value를 출력하고 없으면 key를 value 1000과 함께
# 추가하시오 단, setdefault를 사용하지 마시오..

names = {'park':1500, 'kim':5478, 'minho':8754, 'lee':4589}
nam = input('이름을 입력하세요')

if nam in names:
  print(names[nam])
else:
  names[nam] = 1000
  print(names)
```

```python
# 점수를 입력받은 후,
# 점수가 90이상이면 "A학점입니다"
# 점수가 80이상이면 "B학점입니다"
# 점수가 70이상이면 "C학점입니다"
# 점수가 60이상이면 "D학점입니다"
# 점수가 60미만이면 'F학점입니다'를 출혁하시오

score = int(input('점수를 입력하세요'))

if score >= 90:
  print("A학점입니다")
elif score >= 80:
  print("B학점입니다")
elif score >= 70:
  print("C학점입니다")
elif score >= 60:
  print("D학점입니다")
else:
  print("F학점입니다..")
```

```python
# 임의의 정수 a와 b를 입력받은 후,
# a가 b보다 크면 "a가 b보다 큽니다"
# a가 b보다 작으면 "a가 b보다 작습니다"
# a와 b가 같으면 "a와 b가 같습니다"를 출력하시오..

a = int(input('임의의 정수를 입력하세요'))
b = int(input('임의의 정수를 입력하세요'))

if a>b:
  print('a가 b보다 큽니다')
elif a<b:
  print('a가 b보다 작습니다..')
else:
  print('a와 b가 같습니다..')
```

```python
# 가위, 바위, 보 게임을 if문으로 구현하시오

mine = input('가위, 바위, 보 중에서 하나를 내세요')
yours = input('가위, 바위, 보 중에서 하나를 내세요')

if mine == '가위':
  if yours == '바위':
    print("당신이 이겼습니다")
  elif yours == '보':
    print("내가 이겼습니다")
  else:
    print('비겼습니다')
elif mine == '바위':
  if yours == '보':
    print("당신이 이겼습니다")
  elif yours=='가위':
    print('내가 이겼습니다')
  else:
    print("비겼습니다..")
else:
  if yours == "가위":
    print("당신이 이겼습니다")
  elif yours == '바위':
    print("내가 이겼습니다")
  else:
    print("비겼습니다..")
```
